```python
import time
from textblob import TextBlob
```

```python
list_probs_01 = []
list_len_01 = []

list_probs_02 = []
list_len_02 = []

list_probs_03 = []
list_len_03 = []

list_probs_04 = []
list_len_04 = []

tic = time.clock()
for iddt, strtext in enumerate(df_sales[column]):
    textblob_temp = TextBlob(strtext)
    if len(textblob_temp.words)>=5:
            
        probs = get_n_grams_probability_dict(strtext, 1, unigram_dict)
        list_probs_01.append(probs)
        list_len_01.append(len(probs))

        probs = get_n_grams_probability_dict(strtext, 2, bigram_dict)
        list_probs_02.append(probs)
        list_len_02.append(len(probs))    

        probs = get_n_grams_probability_dict(strtext, 3, trigram_dict)
        list_probs_03.append(probs)
        list_len_03.append(len(probs)) 

        probs = get_n_grams_probability_dict(strtext, 4, fourgram_dict)
        list_probs_04.append(probs)
        list_len_04.append(len(probs)) 
    
    else:
        list_probs_01.append([0])
        list_len_01.append(1)
        list_probs_02.append([0])
        list_len_02.append(1) 
        list_probs_03.append([0])
        list_len_03.append(1)
        list_probs_04.append([0])
        list_len_04.append(1)
    
    if (iddt%1000 == 0): 
        print(iddt, len(df_sales[column]), time.clock() - tic)
```

```python
def get_n_grams_probability(strtext, ngram, ngram_df):
    df_ngram = ngram_analysis.get_n_grams_count(strtext,ngram)
    df_ngram = df_ngram.rename({0: 'ngram', 1: 'count'}, axis='columns')
    df_process = pd.merge(df_ngram, ngram_df, on=["ngram"])
    df_process['probability_mult'] = df_process['probability'] * df_process['count_x']
    mean_prob =  (df_process['probability_mult'].sum() / df_process['count_x'].sum())
    ngram_probs = list(df_process['probability'].quantile([.99, 0.95, 0.9, 0.75, 0.5, 0.25]))
    ngram_probs.append(mean_prob)
    return ngram_probs, df_process
```
