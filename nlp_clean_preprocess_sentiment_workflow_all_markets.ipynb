# All Market Files

```python
files = glob.glob("comments/raw_comments/*.fea")
```

# Filtering by AVM Value / TransPrice Ratio

```python
for file in files:
    print(file)
    df_sales = pd.read_feather(file)
    print(len(df_sales))
    
    '''
    Filtering df_sales
    '''
    df_sales = df_sales[(df_sales['ListingPrice']/df_sales['Transprice'] >= 0.4 ) & 
     (df_sales['ListingPrice']/df_sales['Transprice'] <= 2.5) &
     (df_sales['avmValue']/df_sales['Transprice'] >= 0.4 ) & 
     (df_sales['avmValue']/df_sales['Transprice'] <= 2.5)]    
    filtered_file = file.split('.')[0] + '_filtered.fea'
    df_sales.reset_index().to_feather(filtered_file)
    
```

```python
filtered_files = glob.glob("comments/raw_comments/*_filtered.fea")
```

```python
for file in filtered_files:
    print(file)
    df_sales = pd.read_feather(file)
    df_sales = df_sales.drop(columns='level_0')
    print(len(df_sales))
    
    '''
    Preparing the comments
    '''
    
    tic = time.time()
    df_sales = nlp_pandas_functions.df_column_simple_prepare(df_sales, 'publicremarks', 'publicremarks_prepared')
    toc = time.time()
    print(toc - tic)
    df_sales.to_feather(file)
```
