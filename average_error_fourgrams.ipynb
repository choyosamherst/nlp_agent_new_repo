```python
def get_ngram_errors_dict(strtext, ngram, ngram_dict):
    #tic = time.clock()
    df_ngram = ngram_analysis.get_n_grams_count(strtext,ngram)
    df_ngram = df_ngram.rename({0: 'ngram', 1: 'count'}, axis='columns')
    #print(keys)
    #tic = time.clock()
    #print(dictionary_temp)
    errors = list(map(ngram_dict.get, list(df_ngram['ngram'].values)))
    errors = [0 if v is None else v for v in errors]
    #print(probabilities)
    #print(3, time.clock() - tic)
    
    return errors
```

```python
import time
from textblob import TextBlob
```

```python
list_probs_01 = []
list_len_01 = []

tic = time.clock()
for iddt, strtext in enumerate(df_sales[column]):
    textblob_temp = TextBlob(strtext)
    if len(textblob_temp.words)>=4:
            
        errors = get_ngram_errors_dict(strtext, 4, fourgram_df_dict)
        list_probs_01.append(errors)
        list_len_01.append(len(errors))

    else:
        list_probs_01.append([0])
        list_len_01.append(1)
    
    if (iddt%1000 == 0): 
        print(iddt, len(df_sales[column]), time.clock() - tic)
```

```python
strtext
```

```python
error_array = np.zeros([len(df_sales), np.array(list_len_01).max()])
```

```python
for i in np.arange(len(df_sales)):
    array_tmp = np.array(list_probs_01[i])
    array_tmp = np.concatenate((np.array(list_probs_01[i]), np.zeros(error_array.shape[1]-len(array_tmp))))
    array_tmp = np.sort(array_tmp)
    error_array[i,:] = array_tmp
```

```python
error_df = pd.DataFrame(data=error_array, index=df_sales['index'].values, columns=['fourgram_error_column_'+str(x) for x in range(error_array.shape[1])])
```
