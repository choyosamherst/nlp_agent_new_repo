```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import string

import re

import json
```

```python
with open("replacement_dictionary_pending_currently_processing.json") as json_file:
    replacement_dictionary_2 = json.load(json_file)

```

```python
replacement_dictionary_2
```

```python
df = pd.DataFrame.from_dict(replacement_dictionary_2, orient='index')
df = df.reset_index().rename(columns={'index':'to_replace', 0:'replace_with'})
```

```python
df['string_len'] = df.to_replace.str.len()
```

```python
df = df.sort_values(by=['string_len'], ascending=False)
```

```python
df
```

```python
replacement_dictionary = df[['to_replace', 'replace_with']].set_index('to_replace').to_dict(orient='dict')['replace_with']
```

```python
# create json object from dictionary
json = json.dumps(replacement_dictionary, indent = 4)
f = open("replacement_dictionary_ordered.json","w")
f.write(json)
f.close()
```
