```python
list_probs_01 = []
list_len_01 = []


list_probs_02 = []
list_len_02 = []

list_avmerror = []

tic = time.clock()
for iddt, strtext in enumerate(df_sales[column].values):
    textblob_temp = TextBlob(strtext)
    if len(textblob_temp.words)>=4:
            
        errors = get_noun_sentence_errors_dict(strtext, noun_sentence_dict_01)
        list_probs_01.append(errors)
        list_len_01.append(len(errors))

        errors = get_noun_sentence_errors_dict(strtext, noun_sentence_dict_02)
        list_probs_02.append(errors)
        list_len_02.append(len(errors))   
        
        list_avmerror.append([df_sales['avmerror'].values[iddt]])

    else:
        list_probs_01.append([None])
        list_len_01.append(None)
        list_probs_02.append([None])
        list_len_02.append(None) 
        list_avmerror.append([None])
    
    if (iddt%1000 == 0): 
        print(iddt, len(df_sales[column]), time.clock() - tic)
```
