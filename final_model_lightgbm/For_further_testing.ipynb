```python
def explode(df, lst_cols, fill_value=''):
    # make sure `lst_cols` is a list
    if lst_cols and not isinstance(lst_cols, list):
        lst_cols = [lst_cols]
    # all columns except `lst_cols`
    idx_cols = df.columns.difference(lst_cols)

    # calculate lengths of lists
    lens = df[lst_cols[0]].str.len()

    if (lens > 0).all():
        # ALL lists in cells aren't empty
        return pd.DataFrame({
            col:np.repeat(df[col].values, df[lst_cols[0]].str.len())
            for col in idx_cols
        }).assign(**{col:np.concatenate(df[col].values) for col in lst_cols}) \
          .loc[:, df.columns]
    else:
        # at least one list in cells is empty
        return pd.DataFrame({
            col:np.repeat(df[col].values, df[lst_cols[0]].str.len())
            for col in idx_cols
        }).assign(**{col:np.concatenate(df[col].values) for col in lst_cols}) \
          .append(df.loc[lens==0, idx_cols]).fillna(fill_value) \
          .loc[:, df.columns]
```

```python
for iddd,lid in enumerate(new_avm_df.listingid):
    symptoms_dataframe = pd.DataFrame(data= np.zeros([1,
                                    len(list(noun_sentence_state_train.columns[51:]))]) , 
                    index=[iddd],
                    columns=list(noun_sentence_state_train.columns[51:]))
    print(iddd, len(new_avm_df.listingid))
    set_temp = set(noun_sentence_df_T_exploded[noun_sentence_df_T_exploded.Listingid_List==lid]['index'])
    set_intersected = set_phrases.intersection(set_temp)
    symptoms_dataframe.loc[:,set_intersected] = 1
    
    temp_df = noun_sentence_df_final_reindex_df_filtered.loc[set_intersected]
    
    if len(temp_df)>2:

        temp_df = temp_df.drop(columns=['level_0','AVM_Error_List','Listing_Error_List','Listingid_List'])

        temp_df = temp_df.loc[:, ~temp_df.columns.str.startswith('noun_s')]

        temp_df_mean = noun_sentence_df_final_reindex_df_filtered.loc[set_intersected]
        temp_df_mean = temp_df_mean.mean().to_frame().T
        temp_df_mean.index = [lid]


        temp_df_max_scale =  temp_df.loc\
                [temp_df.AVM_Error_Indicators_scale.idxmax()]

        temp_df_max_value =  temp_df.loc\
                [temp_df.AVM_Error_Indicators_0.idxmax()]

        temp_df_min_value =  temp_df.loc\
                [temp_df.AVM_Error_Indicators_0.idxmin()]

        temp_df_max_scale['AVM_Error_Indicators_0_max_value'] = \
            temp_df_max_value['AVM_Error_Indicators_0']

        temp_df_max_scale['AVM_Error_Indicators_0_min_value'] = \
            temp_df_min_value['AVM_Error_Indicators_0']

        temp_df_max_scale = temp_df_max_scale.to_frame().T
        temp_df_max_scale.index = [lid]

        temp_df_max_indicator  =  temp_df.loc[temp_df.AVM_Error_Indicators_10.idxmax()]
        temp_df_min_indicator  =  temp_df.loc[temp_df.AVM_Error_Indicators_10.idxmin()]

        temp_df_max_scale['AVM_Error_Indicators_0_max_indicator'] = \
            temp_df_max_indicator['AVM_Error_Indicators_10']

        temp_df_max_scale['AVM_Error_Indicators_0_min_indicator'] = \
            temp_df_min_indicator['AVM_Error_Indicators_10']

        temp_df_max_scale['AVM_Error_Indicators_0_diff_indicator'] = \
            temp_df_max_scale['AVM_Error_Indicators_0_max_indicator'] - \
            temp_df_max_scale['AVM_Error_Indicators_0_min_indicator']

        temp_df_max_scale['AVM_Error_Indicators_0_diff_value'] = \
            temp_df_max_scale['AVM_Error_Indicators_0_max_value'] - \
            temp_df_max_scale['AVM_Error_Indicators_0_min_value']

        temp_df_max_scale['AVM_Error_Indicators_0_mean'] =\
            temp_df_mean['AVM_Error_Indicators_0']

        temp_df_max_scale['AVM_Error_Indicators_10_mean'] =\
            temp_df_mean['AVM_Error_Indicators_10']-50.

        for idd,li in enumerate(list_phrases):
            boolean = all((symptoms_dataframe.loc[:, li] == 1).iloc[0].values)
            if(boolean == True): 
                #print(li)
                symptoms_dataframe.loc[:,name_phrases[idd]] = 1

        symptoms_dataframe.index = [lid]
        symptoms_dataframe = pd.merge( \
                                    temp_df_max_scale, \
                                    symptoms_dataframe, \
                                    left_index=True, right_index=True, how='left', 
                                                   suffixes=('_left', '_right'))

        symptoms_dataframe_all = symptoms_dataframe_all.append(symptoms_dataframe)
```
