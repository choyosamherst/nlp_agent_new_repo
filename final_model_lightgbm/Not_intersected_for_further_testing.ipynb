```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
state = 'NC'
```

```python
file_symptoms_train = '../comments/raw_comments/agent_comments_'+state+\
                        '_filtered_final_joined_selected_columns_use_max_scale_one_df_train.fea'

noun_sentence_state_train = pd.read_feather(file_symptoms_train)
```

```python
file_symptoms_test = '../comments/raw_comments/agent_comments_'+state+\
                        '_filtered_final_joined_selected_columns_use_max_scale_one_df_test.fea'
noun_sentence_state_test = pd.read_feather(file_symptoms_test)
```

```python
file = 'NEW_AVM_DATA'+state+'_full.fea'

new_avm_df = pd.read_feather(file)
new_avm_df = new_avm_df.drop(columns=['index','state'])
```

```python
new_avm_df = new_avm_df.rename(columns={'avmvalue':'new_avmValue', 'avmerror':'new_avmerror'})
```

```python
new_avm_df = new_avm_df.set_index('listingid')
```

```python
noun_sentence_state_test = noun_sentence_state_test.set_index('listingid')
noun_sentence_state_train = noun_sentence_state_train.set_index('listingid')
```

```python
old_avmset_train = set(noun_sentence_state_train.index)
old_avmset_test = set(noun_sentence_state_test.index)
```

```python
old_avmset_union = old_avmset_train.union(old_avmset_test)
```

```python
len(old_avmset_union)
```

```python
new_avm_df_set = set(new_avm_df.index)
```

```python
old_new_intersection = old_avmset_union.intersection(new_avm_df_set)
```

```python
new_avm_df = new_avm_df.drop(index=old_new_intersection)
```

```python
len(new_avm_df)
```

```python
file = 'NEW_AVM_DATA'+state+'_not_intersected_for_further_testing.fea'
new_avm_df.reset_index().to_feather(file)
```
