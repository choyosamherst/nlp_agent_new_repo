```python
from gensim.parsing.preprocessing import strip_punctuation 
from gensim.parsing.preprocessing import strip_numeric
from gensim.parsing.preprocessing import strip_short
from gensim.parsing.preprocessing import strip_non_alphanum   
from nltk.stem.porter import PorterStemmer

def clean_jv(doc):
   global countdone   
   p_stemmer = PorterStemmer()
   if countdone == 0: logging.info("Starting to process listings")
   punc_free = strip_punctuation(doc)
   numb_free = strip_numeric(punc_free)
   twol_free = strip_short(numb_free,3)
   splc_free = strip_non_alphanum(twol_free)
   stop_free = " ".join(p_stemmer.stem(lemma.lemmatize(i)) for i in splc_free.lower().split() if i not in stop  and len(i) > 2)
   countdone +=1
   if countdone % 10000 == 0: logging.info("Processed %d listings" %countdone)
   return stop_free
```

```python
from nltk import ngrams

def bigrams_join(n):
    index=0
    for doc in all_docs: 
        bigrams = list(ngrams(doc , n))
        for i in bigrams:
            words = " ".join(i)
            yield [words,df['avmerror'][index],abs(df['avmerror'][index])]
        index+=1
        if index % 10000 == 0: logging.info("Processed %d listings" %index)
```

```python

singlegram_docs = list(bigrams_join(n=1))
singlegram_docs_np=pd.DataFrame(singlegram_docs)
singlegram_docs_np.columns = ['words','err','abserr']

bigram_docs = list(bigrams_join(n=2))
bigram_docs_np=pd.DataFrame(bigram_docs)
bigram_docs_np.columns = ['words','err','abserr']


trigram_docs = list(bigrams_join(n=3))
trigram_docs_np=pd.DataFrame(trigram_docs)
trigram_docs_np.columns = ['words','err','abserr']
```
