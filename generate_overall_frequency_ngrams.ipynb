```python
columns = ['publicremarks', 'clean_publicremarks', 'clean_publicremarks_additional_words',
          'clean_publicremarks_nopunct', 'clean_publicremarks_nopunct_additional_words',
           'clean_publicremarks_stemmed', 'clean_publicremarks_stemmed_additional_words', 
           'clean_publicremarks_lemmatized', 'clean_publicremarks_lemmatized_additional_words'
           ]

for column in columns:
    print(column)
    tic = time.clock()
    unigram_df = ngram_analysis.df_lambda_ngram(df_sales, column, 1)
    bigram_df = ngram_analysis.df_lambda_ngram(df_sales, column, 2)
    trigram_df = ngram_analysis.df_lambda_ngram(df_sales, column, 3)
    fourgram_df = ngram_analysis.df_lambda_ngram(df_sales, column, 4)
    unigram_df = unigram_df.rename({0: 'ngram', 1: 'count'}, axis='columns')
    bigram_df = bigram_df.rename({0: 'ngram', 1: 'count'}, axis='columns')
    trigram_df = trigram_df.rename({0: 'ngram', 1: 'count'}, axis='columns')
    fourgram_df = fourgram_df.rename({0: 'ngram', 1: 'count'}, axis='columns')
    unigram_df['probability'] = unigram_df['count'] / unigram_df['count'].sum()
    bigram_df['probability'] = bigram_df['count'] / bigram_df['count'].sum()
    trigram_df['probability'] = trigram_df['count'] / trigram_df['count'].sum()
    fourgram_df['probability'] = fourgram_df['count'] / fourgram_df['count'].sum()
    unigram_df.to_feather('ngrams/' + file + '_withsentiment_' + column + '_unigram.fea')
    bigram_df.to_feather('ngrams/' + file + '_withsentiment_' + column + '_bigram.fea')
    trigram_df.to_feather('ngrams/' + file + '_withsentiment_' + column + '_trigram.fea')
    fourgram_df.to_feather('ngrams/' + file + '_withsentiment_' + column + '_fourgram.fea')
    toc = time.clock()
    print(toc - tic)
```
